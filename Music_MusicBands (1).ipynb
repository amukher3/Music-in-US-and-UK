```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
res=requests.get('https://www.businessinsider.com/the-100-most-popular-rock-bands-of-all-time-2018-9')
soup=BeautifulSoup(res.text,'html.parser')
```

```python
temp=str(soup.select('div'))
```

```python
import re
tempNames=re.findall('<h2 class=\"slide-title-text\">(\w+.\s*\w+\s*\w+\s*\w+)',temp)
```

```python
tempPrime=(soup.select('span'))
tempTilda=[]
for i in range(37,70,3):
    tempTilda.append(tempPrime[i])
    print(i,tempPrime[i],end='\n')
    
for i in range(125,363,3):
    tempTilda.append(tempPrime[i])
    print(i,tempPrime[i],end='\n')
    
```

```python
len(tempTilda)
```

```python
temp2=[]
tempPrime2=[]
tempTilda2=[]
tempAlpha2=[]
tempAlpha3=[]
for i in range(len(tempTilda)):
    
    tempViews=str(tempTilda[i]).split('Annual Wikipedia page views:')
    tempHits=str(tempTilda[i]).split('Most popular album:')
    tempFormed=str(tempTilda[i]).split('Formed in:')
    tempSales=str(tempTilda[i]).split('All time US album sales:')
    if i==0:
        pass
   #    temp1=tempViews[0].split('>')[1]
    #   temp2.append(temp1.split('<')[0])
        
    if i!=0:
        temp1=tempViews[1].split('>')[1]
        temp2.append(int(temp1.split('<')[0].replace(',','')))
        tempPrime1=tempHits[1].split('>')[1]
        tempPrime2.append(tempPrime1.split('<')[0])
        tempTilda1=tempFormed[1].split('>')[1]
        tempTilda2.append(tempTilda1.split('<')[0])
        tempAlpha1=tempSales[1].split('>')[1]
        tempAlpha2.append(tempAlpha1.split('<')[0])
        tempAlpha3.append(float(tempAlpha2[len(tempAlpha2)-1].split(' million')[0]))

print(temp2)
print(tempAlpha2)
print(tempAlpha3)
print(len(tempAlpha3))
print(len(tempAlpha2))
```

```python
tempLocation=[]
tempYear=[]
for i in tempTilda2:
    tempLocation.append(i.split(' in ')[0])
    tempYear.append(i.split(' in ')[1])
    
print(tempLocation)
print(tempYear)
```

```python
#Distribution of the formed year: 

tempDict=dict()
yearCats={'1960-1965':0,'1966-1970':0,'1971-1975':0,'1976-1980':0,'1981-1985':0,'1986-1990':0,'1991-1995':0,'1996-2000':0,
         '2001-2005':0}

for i in tempYear:
    if i=='early 1970s':
        yearCats['1971-1975']+=1
        continue
    if i=='late 1980s':
        yearCats['1986-1990']+=1
        continue
    if int(i)>=1960 and int(i)<=1965:
        yearCats['1960-1965']+=1
    if int(i)>=1966 and int(i)<=1970:
        yearCats['1966-1970']+=1
    if int(i)>=1971 and int(i)<=1975:
        yearCats['1971-1975']+=1
    if int(i)>=1976 and int(i)<=1980:
        yearCats['1976-1980']+=1
    if int(i)>=1981 and int(i)<=1985:
        yearCats['1981-1985']+=1
    if int(i)>=1986 and int(i)<=1990:
        yearCats['1986-1990']+=1
    if int(i)>=1991 and int(i)<=1995:
        yearCats['1991-1995']+=1
    if int(i)>=1996 and int(i)<=2000:
        yearCats['1996-2000']+=1
    if int(i)>=2001 and int(i)<=2005:
        yearCats['2001-2005']+=1
        
```

```python
sorted(yearCats.items(),key=lambda x:x[1],reverse=True)
```

```python
plt.bar(yearCats.keys(),yearCats.values())
plt.tick_params(axis='x',rotation=70)
```

**According to the data it seems that the period between 1966-1970 had the most number of band formations followed by the period between 1991-1995.***
